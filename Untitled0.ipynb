```python
import random

print("üéØ Welcome to the Number Prediction Game!")
print("I am thinking of a number between 1 and 100.")

number = random.randint(1, 100)
attempts = 0

while True:
    guess = input("Enter your guess (or type 'q' to quit): ")

    if guess.lower() == 'q':
        print("Game exited. The number was:", number)
        break

    if not guess.isdigit():
        print("‚ùå Please enter a valid number.")
        continue

    guess = int(guess)
    attempts += 1

    if guess < number:
        print("‚¨Ü Too low! Try again.")
    elif guess > number:
        print("‚¨á Too high! Try again.")
    else:
        print(f"üéâ Correct! You guessed the number in {attempts} attempts.")
        break
```
